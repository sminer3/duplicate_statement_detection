```python
import pandas as pd
import numpy as np
```

```python
orig = pd.read_csv('../data/train.csv')
cleanDat = pd.read_csv('../data/train_splits_removed.csv')
```

```python
cleanDat.head()
```

```python
orig.head()
```

```python
orig_statements = list(orig.statement1.apply(str)) + list(orig.statement2.apply(str))
length_statements = [len(s) for s in orig_statements]
print('Average number of characters per statement in the original data', np.round(np.mean(length_statements),2))
clean_statements = list(cleanDat.statement1.apply(str)) + list(cleanDat.statement2.apply(str))
length_statements_clean = [len(s) for s in clean_statements]
print('Average number of characters per statement in the filtered data', np.round(np.mean(length_statements_clean),2))
```

```python
#Sample of statement pairs that are in the filtered data:
import random
pd.set_option('display.max_colwidth', -1)
samp = random.sample(range(len(cleanDat)), 10)
cleanDat.iloc[samp,:].loc[:,['statement1','statement2']]
```

```python
#Sample of statement pairs that are in the filtered data:
loss_dat = orig[~orig.id_all.isin(cleanDat.id_all)]
samp = random.sample(range(len(loss_dat)), 10)
loss_dat.iloc[samp,:].loc[:,['statement1','statement2']]
```
